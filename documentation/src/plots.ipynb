```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the current implementation (V2 or whatever you are plotting)
N_exponents = [20, 23, 25, 28]  # Exponents of 2
execution_time = [20.121857, 179.788803, 799.303894, 7313.434570]

# Data for MPI (adjust these values accordingly)
mpi_execution_time = [111.0, 1242.234567, 6277.987654, 127953.29841]  # Placeholder MPI data

# Calculate N values as 2^exponent
N_values = [2 ** n for n in N_exponents]

# Convert exponents into labels like 2^20, 2^23, etc.
N_labels = [f'$2^{{{n}}}$' for n in N_exponents]

# Create the plot with Execution Time on the vertical axis and N on the horizontal axis
plt.figure(figsize=(8, 5))

# Plot your current algorithm's execution time
plt.plot(N_values, execution_time, marker='o', linestyle='-', color='b', label='Current Algorithm')

# Plot MPI execution time
plt.plot(N_values, mpi_execution_time, marker='x', linestyle='--', color='r', label='MPI Algorithm (p = 4)')

# Set logarithmic scale for both axes
plt.xscale('log')
plt.yscale('log')

# Customize the x-axis with 2^ labels
plt.xticks(N_values, N_labels)

# Add labels and title
plt.xlabel('Size of Array', rotation=0, labelpad=20)  # Rotate 'N' for clarity
plt.ylabel('Execution Time (ms)')
plt.title('Execution Time vs N (Logarithmic Scale for Both Axes)')
plt.grid(True, which="both", linestyle='--', linewidth=0.5)

# Modify the y-ticks to show execution time values
yticks = plt.gca().get_yticks()

# Annotate execution times on the y-axis
ytick_labels = [f'{tick:.2f} ms' for tick in yticks]

# Update the y-ticks with the execution time labels
plt.yticks(yticks, ytick_labels)

# Add a legend to distinguish between the two datasets
plt.legend()

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt

# Data for the pillar chart
categories = ['V0', 'V1', 'V2', 'p1', 'p2', 'p4', 'p8']
execution_times = [4814.9292, 435.678223, 20.121857, 152, 114, 111, 100]

# Creating the pillar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, execution_times, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Execution Time (ms)', fontsize=12)
plt.title('Execution Time by Category', fontsize=14)
plt.yscale('log')  # Using a logarithmic scale for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding the exact execution time values on top of each pillar
for bar, time in zip(bars, execution_times):
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # X position
        bar.get_height(),  # Y position
        f'{time:.3f}',  # Text to display (rounded to 3 decimal places)
        ha='center', va='bottom', fontsize=10  # Centered alignment
    )

# Display the chart
plt.tight_layout()
plt.show()
```
